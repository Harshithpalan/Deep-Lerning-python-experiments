```python
# Download an example image if it doesn't exist
image_url = "https://ultralytics.com/images/zidane.jpg"
image_filename = "image.jpg"

try:
    with open(image_filename, 'rb') as f:
        pass
except FileNotFoundError:
    print(f"Downloading {image_url} to {image_filename}...")
    urllib.request.urlretrieve(image_url, image_filename)
    print("Download complete.")
```

```python
import cv2
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)

    annotated_frame = results[0].plot()

    cv2.imshow("YOLO Detection", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO
import urllib.request
import os
from IPython.display import display, Image
import numpy as np

video_url = "https://ultralytics.com/assets/bus.mp4" # Corrected video URL
video_filename = "bus.mp4"

if not os.path.exists(video_filename):
    print(f"Downloading {video_url} to {video_filename}...")
    urllib.request.urlretrieve(video_url, video_filename)
    print("Download complete.")
else:
    print(f"{video_filename} already exists. Skipping download.")
```

```python
# Open the video file
cap = cv2.VideoCapture(video_filename)

if not cap.isOpened():
    print(f"Error: Could not open video file {video_filename}.")
else:
    # Get video properties
    fps = cap.get(cv2.CAP_PROP_FPS)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # Define the codec and create VideoWriter object
    output_filename = "output_video_yolov8m.mp4" # Changed output filename to differentiate
    # Use MP4V for .mp4 output
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_filename, fourcc, fps, (width, height))

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Run YOLO detection on the frame
        results = model(frame, verbose=False) # verbose=False to suppress per-frame output

        # Annotate the frame
        annotated_frame = results[0].plot()

        # Write the annotated frame to the output video
        out.write(annotated_frame)

    # Release everything when job is finished
    cap.release()
    out.release()
    cv2.destroyAllWindows()
    print(f"Processing complete. Output video saved as {output_filename}")
    print("You can download 'output_video_yolov8m.mp4' from the files pane to view the results.")

model = YOLO("yolov8m.pt")

results = model('image.jpg', show=True)
results[0].save('output_m_model.jpg')
print("Detection with yolov8m.pt complete. Output saved as output_m_model.jpg")

from IPython.display import Image
Image('output_m_model.jpg')
```
