```python
from transformers import (
    pipeline,
    AutoTokenizer,
    AutoModelForSeq2SeqLM
)
```

1️⃣ TEXT GENERATION (GPT-style)

```python
def generate_text(prompt):
    generator = pipeline("text-generation", model="gpt2")
    result = generator(prompt, max_length=100, num_return_sequences=1)
    return result[0]['generated_text']
```

2️⃣ QUESTION ANSWERING (BERT)

```python
def answer_question(context, question):
    qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
    result = qa_pipeline(question=question, context=context)
    return result['answer']
```

3️⃣ LANGUAGE TRANSLATION

```python
def translate_text(text, src_lang="en", tgt_lang="fr"):
    model_name = f"Helsinki-NLP/opus-mt-{src_lang}-{tgt_lang}"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

    inputs = tokenizer(text, return_tensors="pt", padding=True)
    translated = model.generate(**inputs)
    return tokenizer.decode(translated[0], skip_special_tokens=True)
```

4️⃣ TEXT SUMMARIZATION

```python
def summarize_text(text):
    model_name = "facebook/bart-large-cnn"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

    inputs = tokenizer(text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(inputs["input_ids"], num_beams=4, max_length=100, min_length=30, early_stopping=True)
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```

 MAIN MENU

```python
def main():
    print("\n--- Transformer NLP Application ---")
    print("1. Text Generation (GPT)")
    print("2. Question Answering (BERT)")
    print("3. Translation")
    print("4. Summarization")

    choice = input("Select task (1-4): ")

    if choice == "1":
        prompt = input("Enter prompt: ")
        print("\nGenerated Text:\n", generate_text(prompt))

    elif choice == "2":
        context = input("Enter context: ")
        question = input("Enter question: ")
        print("\nAnswer:\n", answer_question(context, question))

    elif choice == "3":
        text = input("Enter text: ")
        src = input("Source language code (e.g., en): ")
        tgt = input("Target language code (e.g., fr): ")
        print("\nTranslated Text:\n", translate_text(text, src, tgt))

    elif choice == "4":
        text = input("Enter text to summarize: ")
        print("\nSummary:\n", summarize_text(text))

    else:
        print("Invalid choice!")

if __name__ == "__main__":
    main()
```
